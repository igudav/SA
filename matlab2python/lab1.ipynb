### Task 5

```python
def isprime(n: int):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n) + 1)):
        if n % i == 0:
            return False
    return True

EPS = 1e-6
NSAMPLES = 100
n = int(input('n: '))
if not isprime(n):
    print('not prime')
else:
    err_solve = np.zeros(NSAMPLES)
    err_inv = np.zeros(NSAMPLES)
    err_pinv = np.zeros(NSAMPLES)
    for i in range(0, NSAMPLES):
        A = np.random.rand(n, n)
        b = np.random.rand(n, 1)
        
        if (np.linalg.det(A) > EPS):
            x = np.linalg.solve(A, b)
            err_solve[i] = np.linalg.norm(A @ x - b)
            
            x = np.linalg.inv(A) @ b
            err_inv[i] = np.linalg.norm(A @ x - b)
            
            x = np.linalg.pinv(A) @ b
            err_pinv[i] = np.linalg.norm(A @ x - b)
        else:
            i -= 1
    
    fig, ax = plt.subplots()
    ax.plot(err_solve, label='solve')
    ax.plot(err_inv, label='inv')
    ax.plot(err_pinv, label='pinv')
    ax.set_ylabel('error')
    ax.legend()
```

### Task 9

```python
def my_multiply(A, B):
    if A.shape[1] != B.shape[0]:
        raise 0
        return -1
    C = np.zeros((A.shape[0], B.shape[1]))
    for i in range(0, A.shape[0]):
        for j in range(0, B.shape[1]):
            for k in range(0, A.shape[1]):
                C[i][j] += A[i][k] * B[k][j]
    return C

MAX_DIM_EXP = 1.5
NGRID = 40
NREPS = 30

dims = np.logspace(1, MAX_DIM_EXP, NGRID)
py_times = np.zeros(NGRID)
my_times = np.zeros(NGRID)
for i in range(0, NGRID):
    A = np.random.rand(int(np.fix(dims[i])), int(np.fix(dims[i])))
    B = np.random.rand(int(np.fix(dims[i])), int(np.fix(dims[i])))
    tmp_my_times = np.zeros(NREPS)
    tmp_py_times = np.zeros(NREPS)
    for j in range(0, NREPS):
        tmp_my_times[j] = time.time_ns()
        my_multiply(A, B)
        tmp_my_times[j] = time.time_ns() - tmp_my_times[j]
        tmp_py_times[j] = time.time_ns()
        C = A @ B
        tmp_py_times[j] = time.time_ns() - tmp_py_times[j]
    my_times[i] = np.median(tmp_my_times)
    py_times[i] = np.median(tmp_py_times)

fig, ax = plt.subplots()
ax.plot(dims, py_times / 1e9, label=r'A @ B')
ax.plot(dims, my_times / 1e9, label=r'my_multiply(A, B)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('n')
ax.set_ylabel('t')
ax.set_title('time of multiplication')
ax.legend(loc='upper left')
```

### Task 12

#### Functions

```python
def rectangles(y, x=None, dx=1.0, axis=-1):
    if (x is None):
        return dx * np.sum(y, axis)
    else:
        return np.sum(np.diff(x, axis=axis) * np.take(y, np.arange(y.shape[axis] - 1), axis), axis)
```

```python
def simpson(y, dx=1.0, axis=-1):
    m = y.shape[axis]
    if np.mod(m, 2) != 1:
        raise 'bad num of elems'
    ind2 = np.arange(2, m - 2, 2)
    ind4 = np.arange(1, m - 1, 2)
    return dx / 3 * (np.take(y, 0, axis=axis) + np.take(y, m - 1, axis=axis) + 
            2 * np.sum(np.take(y, ind2, axis=axis), axis=axis) + 4 * np.sum(np.take(y, ind4, axis=axis), axis=axis))
```

#### Setting constants

```python
REP = 20
FROM = 0
TO = 15
MEAS_NUM_STEPS = 20
NGRID = 101
```

#### Plotting antiderivative

```python
x = np.linspace(FROM, TO, NGRID)
dx = x[1] - x[0]
X = np.tile(x, (NGRID, 1))
Y = np.tril(np.sin(X) / X)
Y[np.isnan(Y)] = 1
yr = rectangles(Y, dx=dx)
yt = np.trapz(Y, dx=dx)
ys = simpson(Y, dx=dx)

fig, ax = plt.subplots()
ax.plot(x, yr, label='rectangles')
ax.plot(x, yt, label='trapz', ls='--')
ax.plot(x, ys, label='simpson', ls=':')
ax.legend()
ax.set_title(r'$y=\int_0^x \frac{\sin(t)}{t}\,dt$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
```

#### Measuring errors

```python
res_r = np.zeros(MEAS_NUM_STEPS)
res_t = np.zeros(MEAS_NUM_STEPS)
res_s = np.zeros(MEAS_NUM_STEPS)
steps = np.cumprod(0.5 * np.ones(MEAS_NUM_STEPS))

for i in range(MEAS_NUM_STEPS):
    x = np.arange(FROM, TO + steps[i], steps[i])
    y = np.sin(x) / x
    y[np.isnan(y)] = 1
    res_r[i] = rectangles(y, dx=steps[i])
    res_t[i] = np.trapz(y, dx=steps[i])
    res_s[i] = simpson(y, dx=steps[i])

fig, ax = plt.subplots()
ax.plot(steps[1:], np.abs(np.diff(res_r)), label='rectangles')
ax.plot(steps[1:], np.abs(np.diff(res_t)), label='np.trapz', ls='--')
ax.plot(steps[1:], np.abs(np.diff(res_s)), label='simpson', ls=':')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('error')
ax.legend()
```
